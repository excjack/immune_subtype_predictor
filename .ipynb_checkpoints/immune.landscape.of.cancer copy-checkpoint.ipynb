# Aligning Gene Sets to their Ensembl ID's

```python
ensg = pd.read_csv("hugo2ensembl.tsv", delimiter='\t')
ensg #the holy grail - has all ensembl id's
```

```python
panimmune = pd.read_excel("panimmune.xlsx", sheet_name='Genes')
panimmune #the toolkit - has the diffrent gene sets that 
```

```python
duplicate_values_count = panimmune['SetName'].duplicated(keep=False).sum()

print("Number of duplicates:", duplicate_values_count) 
#tells us how many duplicate values there are. Take away the result with the OG, and youll get number of unique
```

```python
duplicates_mask = panimmune['SetName'].duplicated(keep=False)

df_duplicates = panimmune[duplicates_mask]

names_with_duplicates = df_duplicates['SetName'].unique()

print("List of names with duplicates:")
print(names_with_duplicates)
#tells us what all the set combos are called
```

# Selecting a gene set to filter our big dataset with

```python
value_to_check = 'IR7_score'

first_test_ds = panimmune[panimmune['SetName'] == value_to_check]

first_test_ds  #picking out a single set column and assigning it a varible name
```

```python
first_test_df = pd.merge(ensg, first_test_ds, left_on = 'symbol', right_on = 'Gene', how = 'inner')
first_test_df #making first testing ensembl filter
```

```python
first_test_filter = first_test_df.iloc[:, 1:2]
first_test_filter
```

```python
merged
```

# Applying Filter (for 1st test filter: IR7_score)

```python
ensemblIDs = set(first_test_filter['id'])
cols = list(merged.columns)
newCols = [a for a in cols if a in ensemblIDs]
finalCols = cols[0:3]
finalCols.extend(newCols)

full_first_test_df = merged[finalCols]
```

```python
full_first_test_df
```

# Getting Ready To Train The Machine 
### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(full_first_test_df, test_size=0.25)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
df_train.head()
```

# Prepping Data For First Set

```python
x_train = df_train.drop(columns=["Immune Subtype", "TCGA Participant Barcode"]).values #we drop the columns we dont want to use to train, and we turn the graph back into arrays
y_train = df_train["Immune Subtype"].values
```

```python
x_train
```

```python
y_train
```

```python
x_train.shape
```

```python
df_train
```

```python
xt, xv, yt, yv = train_test_split(x_train, y_train, test_size=0.25)
```

# Modeling - Random Forest Generator Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
rfModel = RandomForestClassifier()
```

```python
rfModel.fit(X=xt, y=yt)
```

```python
#testing accuracy for training set
rfModel.score(x_train, y_train)
```

```python
rfModel.predict(xv)
```

```python
y_pred = rfModel.predict(xv)
```

```python
y_pred == yv
```

```python
rfModel.score(xv, yv) #accuracy score for testing set
```

```python
np.mean(y_pred == yv)
```

# Cross Set Validation for rfModel

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
cv_rfModel = RandomForestClassifier()
```

```python
cross_val_score(cv_rfModel, x_train, y_train, cv=5, scoring='accuracy')
```

```python
accuracies = cross_val_score(cv_rfModel, x_train, y_train, cv=5, scoring='accuracy')
np.mean(accuracies)
```
